```python
#Create a list of numbers from 1 to 10 and print only the elements at even indices
number = list(range(1, 11))
even_index = number[::2]
print(number)
print(f'Elements and even indices are: {even_index}')

```

```python
#Check whether a number is positive, negative, or zero using if-elif-else.
my_numbers = int(input('Enter a number!: '))
if my_numbers > 0:
  print('Positive')
elif my_numbers < 0:
  print('Negative')
else:
  print(my_numbers)
```

```python
#Use a while loop to print numbers from 10 to 1
count = 10
while count >= 1:
    print(count, end = " ")
    count -= 1
print()
```

```python
#Find the sum of all elements in a list using a loop
my_number = [3, 5, 1, 4, 5,7,6]
total = 0
for num in my_number:
    total += num

print(total)
```

```python
#Print all numbers divisible by 3 from a list
nancy =[23, 45, 3, 27, 56, 77, 36]
for i in nancy:
  if i % 3 == 0:
    print(i, end = " ")

```

```python
#Use a loop and conditional statements to find the largest number in a list
data_points = [45, 82, 19, 94, 31]
largest = data_points[0]
for n in data_points:
    if n > largest:
        largest = n
print(f"Largest number: {largest}")
```
